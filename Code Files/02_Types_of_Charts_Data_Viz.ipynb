```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.set(style = 'whitegrid')

#----Data-Decides---by---Omkar-Pawar----#

sorted_df = df.groupby('product_category')['quantity'].sum().reset_index().sort_values(by='quantity', ascending=False)

sns.barplot(x='product_category', y='quantity', data=sorted_df, estimator=sum, ci=None)

#------like---share---subscribe------- -#

plt.title('Total Quantity Sold per Product Category Ordered')
plt.xlabel('Product Category')
plt.ylabel('Total Quantity Sold')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.set(style = 'whitegrid')

#----Data-Decides---by---Omkar-Pawar----#

df['order_date'] = pd.to_datetime(df['order_date'])
daily_orders = df.groupby('order_date')['quantity'].sum().reset_index()

sns.barplot(x='order_date', y='quantity', data=daily_orders)

#------like---share---subscribe------- -#

plt.title('Total Quantity Sold per Day')
plt.xlabel('Date')
plt.xticks(rotation=90)  
plt.ylabel('Total Quantity Sold')
plt.show()
```

### Pie Chart

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.set(style = 'whitegrid')
#----Data-Decides---by---Omkar-Pawar----#

category_counts = df.groupby('product_category')['quantity'].sum().reset_index()

plt.pie(category_counts['quantity'],  
labels=category_counts['product_category'], 
autopct='%1.1f%%', startangle=90, 
colors=sns.color_palette("Set3"))

#------like---share---subscribe---------#
plt.title('Product Distribution by Category')
plt.show()
```

### Histogram

```python
import pandas as pd
file_path = "/home/omkar/Omkar Pawar/Data Science/Data Decides Code for YouTube Shorts/Datasets/ecommerce_online_shop.csv"
df = pd.read_csv(file_path)
```

```python
df[['CustomerID', 'Name', 'Age', 'product_category']].head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.set(style = 'whitegrid')

#----Data-Decides---by---Omkar-Pawar----#

sorted_df = df.groupby('product_category')['quantity'].sum().reset_index().sort_values(by='quantity',
                        ascending=False)

sns.barplot(x='product_category', y='quantity', 
            data=sorted_df, estimator=sum, errorbar=None)

#------like---share---subscribe------- -#

plt.title('Total Quantity Sold per Product Category Ordered')
plt.xlabel('Product Category')
plt.ylabel('Total Quantity Sold')
plt.show()
```

```python
# 1 bar chart for Age
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.set(style = 'whitegrid')

#----Data-Decides---by---Omkar-Pawar----#

sns.barplot(data=df, x='Age', y=df['Age'].value_counts(), 
            errorbar=None)

#------like---share---subscribe------- -#

plt.title('Bar chart')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2 histogram for Age
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.set(style = 'whitegrid')

#----Data-Decides---by---Omkar-Pawar----#

sns.histplot(data=df, x='Age', bins=5, kde=True, 
            color='skyblue', edgecolor='black')

#------like---share---subscribe---------#

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram ')
plt.show()
```
